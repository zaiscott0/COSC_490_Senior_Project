**Tax Base Value**

The assessed value of a property for tax purposes . The Tax Base Value is determined by the tax assessor's office and is used to calculate property taxes owed by the property owner.

**Base Full Cash Land Value**

The Land Value refers to the assessed value of the land portion of the property. This value is determined separately from the Improvement Value, and it takes into account factors such as the size and location of the land, the availability of utilities and services, and any zoning restrictions or other regulations that may affect the value of the land.

**Base Full Cash Improvement**

The Improvement Value refers to the assessed value of any improvements made to the property, such as buildings, structures, or other improvements. The Improvement Value is determined by the tax assessor's office and is based on factors such as the size, quality, and age of the improvements.



```python
#to visualize and relate some of these attributes, I create a scatter plot to show the 
#relationship between the Tax Base Value and the Base Full Cash Land Value and Improvement Value. This will allow us to 
#see how the assessed value of the land and improvement value contribute to the overall Tax Base Value of a property.

import pandas as pd
import matplotlib.pyplot as plt


# Select the relevant columns
data = df[['TAXBASE', 'BFCVLAND', 'BFCVIMPR']]

# Create a scatter plot
plt.scatter(data['BFCVLAND'], data['TAXBASE'], label='Land Value')
plt.scatter(data['BFCVIMPR'], data['TAXBASE'], label='Improvement Value')

# Set axis labels and title
plt.xlabel('Assessed Value')
plt.ylabel('Tax Base Value')
plt.title('Relationship between Assessed Value and Tax Base Value')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

**Struct Area**

**Definition**.

 The sum of all areas on all floors of a building that cannot be occupied or put to use because of structural building features.

**Basis for Measurement**

Precise computation by direct measurement is not possible under this definition. It is determined by calculating the difference between the measured gross area and the measured net usable area.

**Measured in terms of area,**

Structural Area = gross area - Net Usable Area

**Description**

Examples of building features normally classified as Structural Areas include exterior walls, fire walls, and permanent partitions, unusable areas in attics or basements, or comparable portions of a building with ceiling height restrictions.
